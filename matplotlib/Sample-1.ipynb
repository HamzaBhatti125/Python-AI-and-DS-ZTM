```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#first method
x = [1,2,3,4]
y=[11,22,33,44]
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
#2nd Method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y) #add some data
plt.show()
```

```python
#3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x,y)
```

## MAtplotlib example workflow

```python
x = [1,2,3,4]
y= [11,22,33,44]

fig, ax = plt.subplots(figsize=(10,10))

ax.plot(x,y)

ax.set(title="Simple Plt", 
      xlabel="x-axis",
      ylabel="y-axis")

fig.savefig("./images/sample-1.png")
```

```python
#Plot from Dictionary
butter_prices = {"Almond Butter": 10, 
                "Peanut Buter": 8,
                "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values())
ax.set(title="Butter Store");
```

```python
fig, ax = plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()));
```

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
#Option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
ax1.plot(x,x/2)
ax2.scatter(np.random.rand(10), np.random.rand(10));
ax3.bar(butter_prices.keys(), butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#plot to each different index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.rand(10), np.random.rand(10));
ax[1,0].bar(butter_prices.keys(), butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
heart_disease = pd.read_csv('./data/13.1 heart-disease.csv')
heart_disease
```

```python
#histogram
heart_disease["age"].plot.hist();
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
#Pyplot method
over_50.plot(kind="scatter", 
            x="age",
            y="chol",
            c="target");
```

```python
#Oo method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
#ax.set_xlim([45,100]); #change limitations

ax.set(title="Heart Disease", xlabel="Age", ylabel="Cholestrol");

```

```python
#Oo method
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
#ax.set_xlim([45,100]); #change limitations

ax.set(title="Heart Disease", xlabel="Age", ylabel="Cholestrol");

ax.legend(*scatter.legend_elements(), title="Target");

#add horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--");

```
